````kotlin
val descFile = File("/Users/nicopico/Downloads/DeviceInfoPush.desc")
val descriptorSet = DescriptorProtos.FileDescriptorSet.parseFrom(descFile.inputStream())
descriptorSet
````

````kotlin
import com.google.protobuf.Descriptors

val messageTypeDescriptor = descriptorSet.fileList
    .map { Descriptors.FileDescriptor.buildFrom(it, emptyArray()) }
    .flatMap { it.messageTypes }
    .firstOrNull { it.name == "DeviceInfoPush" }
messageTypeDescriptor
````

````kotlin
import com.google.protobuf.DynamicMessage

val payload: String = "0A0568656C6C6F1001"
val dynamicMessage = DynamicMessage.parseFrom(messageTypeDescriptor, payload.hexToByteArray())
dynamicMessage
````

````kotlin
import com.google.protobuf.util.JsonFormat

JsonFormat.printer().print(dynamicMessage)
````
